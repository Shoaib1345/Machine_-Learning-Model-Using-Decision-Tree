```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
label_encoder=LabelEncoder()
#Encode the species  column
read['species']=label_encoder.fit_transform(read['species'])
#Display the first five rows to see the encoding
read.head()
```

```python
#Separating the Target variable(Dependent) 
#Define features and target variable
x=read.drop('species',axis=1)
y=read['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
read.head()
```

```python

read['species'].value_counts()
```

```python
#Creating a Decision Tree Classifier
#initialize the Decision Tree Classifier
dt_classifier=DecisionTreeClassifier(random_state=42)

dt_classifier.fit(x_train , y_train)

DecisionTreeClassifier(random_state=42)
```

```python
x_test.head()
```

```python
x_test.shape
```

```python
y_predicate=dt_classifier.predict(x_test) 
```

```python
y_predicate
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test,y_predicate)
print(f'Accuracy: {accuracy:.2f}')

# Print classification report
print('Classification Report:')
print(classification_report(y_test,y_predicate))

# Print confusion matrix
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_predicate))
```
